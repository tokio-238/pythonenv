```python
%matplotlib inline
from IPython import display
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import mglearn
from IPython.display import display
from sklearn.model_selection import train_test_split

```

```python
mglearn.plots.plot_scaling()
```

```python
from sklearn.datasets import load_breast_cancer
cancer =  load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(cancer.data,cancer.target,random_state=1)

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
#データを変換
X_train_scaled = scaler.transform(X_train)
#スケール変換の前後のデータ特性をプリント
print("transformed shape:{}".format(X_train_scaled.shape))
print("per-feature minimum before scaling: shape:\n{}".format(X_train.min(axis=0)))
print("per-feature maximum before scaling: shape:\n{}".format(X_train.max(axis=0)))
print("per-feature minimum after scaling: shape:\n{}".format(X_train_scaled.min(axis=0)))
print("per-feature maximum after scaling: shape:\n{}".format(X_train_scaled.max(axis=0)))
```

```python
#データを変換
X_test_scaled = scaler.transform(X_test)
#スケール変換の前後のデータ特性をプリント
print("per-feature minimum after scaling: shape:\n{}".format(X_test_scaled.min(axis=0)))
print("per-feature maximum after scaling: shape:\n{}".format(X_test_scaled.max(axis=0)))
```

```python
from sklearn.datasets import make_blobs
#合成データを作成
X, _ = make_blobs(n_samples=50, centers=5, random_state= 4, cluster_std=2)
#訓練セットとデータセットに分割
X_train, X_test = train_test_split(X,random_state=5, test_size=.1)

#訓練セットとテストセットをプロット
fig, axes = plt.subplots(1,3, figsize=(13,4))
axes[0].scatter(X_train[:,0],X_train[:,1], c= mglearn.cm2(0),label = "Training set", s= 60)
axes[0].scatter(X_test[:,0],X_test[:,1],marker="^",c= mglearn.cm2(1),label = "Test set", s= 60)
axes[0].legend(loc = 'upper left')
axes[0].set_title("Original Data")

#MinMax Scalerでデータをスケール変換
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

#スケール変換されたデータの特性を可視化
axes[1].scatter(X_train_scaled[:,0],X_train_scaled[:,1], c= mglearn.cm2(0),label = "Training set", s= 60)
axes[1].scatter(X_test_scaled[:,0],X_test_scaled[:,1],marker="^",c= mglearn.cm2(1),label = "Test set", s= 60)
axes[1].legend(loc = 'upper left')
axes[1].set_title("Scaled Data")

#テストセットを訓練セットとは別にスケール変換
#最小値と最大値が0,1になる
#*実際にはやってはいけない*
test_scaler = MinMaxScaler()
test_scaler.fit(X_test)
X_test_scaled_badly = test_scaler.transform(X_test)

#間違ってスケール変換されたデータを可視化
axes[2].scatter(X_train_scaled[:,0],X_train_scaled[:,1], c= mglearn.cm2(0),label = "Training set", s= 60)
axes[2].scatter(X_test_scaled_badly[:,0],X_test_scaled_badly[:,1],marker="^",c= mglearn.cm2(1),label = "Test set", s= 60)
axes[2].legend(loc = 'upper left')
axes[2].set_title("Improperly Scaled Data")

for ax in axes:
    ax.set_xlabel("Feature 0")
    ax.set_ylabel("Feature 1")
```
